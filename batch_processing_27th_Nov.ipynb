```python
import pandas as pd
import numpy as np
from typing import List, Optional
import psutil
import time
from datetime import datetime

def process_large_dataset(
    df: pd.DataFrame,
    career_column: str = 'Getting Started',
    label_column: str = 'label',
    text_column: str = 'text',
    batch_size: int = 10000
) -> pd.DataFrame:
    """
    Efficiently process large datasets by combining text for each career and label combination.

    Parameters:
    -----------
    df : pandas.DataFrame
        Input DataFrame containing the data
    career_column : str
        Name of the column containing career information
    label_column : str
        Name of the column containing label information
    text_column : str
        Name of the column containing text to be combined
    batch_size : int
        Size of batches for processing

    Returns:
    --------
    pandas.DataFrame
        Processed DataFrame with combined text for each career-label combination
    """

    # Step 1: Clean the input data
    df = df.copy()
    df[text_column] = df[text_column].fillna('')  # Handle NaN values

    # Step 2: Create a compound key for grouping
    df['_group_key'] = df[career_column] + '|||' + df[label_column]

    # Step 3: Process in batches
    result_chunks = []

    for chunk_start in range(0, len(df), batch_size):
        chunk_end = min(chunk_start + batch_size, len(df))
        chunk = df.iloc[chunk_start:chunk_end]

        # Group and aggregate the chunk
        grouped_chunk = (chunk
            .groupby('_group_key', as_index=False)
            .agg({
                text_column: lambda x: ' '.join(x),
                career_column: 'first',
                label_column: 'first'
            })
        )

        result_chunks.append(grouped_chunk)

    # Step 4: Combine all chunks
    result_df = pd.concat(result_chunks, ignore_index=True)

    # Step 5: Merge overlapping groups across chunks
    final_result = (result_df
        .groupby('_group_key', as_index=False)
        .agg({
            text_column: lambda x: ' '.join(x),
            career_column: 'first',
            label_column: 'first'
        })
    )

    # Step 6: Clean up and prepare final output
    final_result = final_result.drop('_group_key', axis=1)
    final_result = final_result.rename(columns={text_column: 'combined_text'})

    return final_result

def process_with_monitoring(
    df: pd.DataFrame,
    career_column: str = 'Getting Started',
    label_column: str = 'label',
    text_column: str = 'text',
    batch_size: int = 10000
) -> pd.DataFrame:
    """
    Wrapper function that includes progress monitoring and memory usage tracking.
    """


    start_time = time.time()
    initial_memory = psutil.Process().memory_info().rss / 1024 / 1024  # MB

    print(f"Starting processing at {datetime.now()}")
    print(f"Initial memory usage: {initial_memory:.2f} MB")
    print(f"Input DataFrame shape: {df.shape}")

    result = process_large_dataset(
        df=df,
        career_column=career_column,
        label_column=label_column,
        text_column=text_column,
        batch_size=batch_size
    )

    end_time = time.time()
    final_memory = psutil.Process().memory_info().rss / 1024 / 1024  # MB

    print(f"\nProcessing completed at {datetime.now()}")
    print(f"Final memory usage: {final_memory:.2f} MB")
    print(f"Memory change: {final_memory - initial_memory:.2f} MB")
    print(f"Total processing time: {end_time - start_time:.2f} seconds")
    print(f"Output DataFrame shape: {result.shape}")

    return result
```

```python
import openai

from openai import OpenAI
# Set your OpenAI API key
openai.api_key = 'your_openai_api_key'

client = OpenAI()

def process_text_with_openai(df: pd.DataFrame, text_column: str = 'combined_text') -> pd.DataFrame:
    """
    Process the text column of the DataFrame using OpenAI API.

    Parameters:
    -----------
    df : pandas.DataFrame
        Input DataFrame with a text column
    text_column : str
        Name of the text column to be sent to OpenAI API

    Returns:
    --------
    pandas.DataFrame
        DataFrame with an additional column for OpenAI API responses
    """
    responses = []

    print(f"Sending text to OpenAI API for processing. Total rows: {len(df)}")

    for i, text in enumerate(df[text_column]):
        try:
            # Sending the text to OpenAI API
            response = client.chat.completions.create(
              model="gpt-4o-mini",
              temperature=0.2,
                messages=[
                    {
                        "role": "user",
                        "content": "Write a haiku about recursion in programming."
                    }
                ],
              response_format={ "type": "json_object" }
              )
            # Extract the response content
            processed_text = response.choices[0].text.strip()
            responses.append(processed_text)
        except Exception as e:
            print(f"Error processing row {i}: {e}")
            responses.append(None)

    # Add the responses to a new column in the DataFrame
    df['openai_response'] = responses
    return df

# Example usage
if __name__ == "__main__":
    import pandas as pd

    # Load your CSV file
    input_file = "your_input_file.csv"
    df = pd.read_csv(input_file)

    # Process the data using the provided functions
    processed_df = process_with_monitoring(df, career_column='Getting Started', label_column='label', text_column='text', batch_size=10000)

    # Send the combined text to OpenAI API
    final_df = process_text_with_openai(processed_df, text_column='combined_text')

    # Save the final DataFrame to a new CSV file
    output_file = "processed_output_with_openai.csv"
    final_df.to_csv(output_file, index=False)

    print(f"Processing completed. Output saved to {output_file}")
```
